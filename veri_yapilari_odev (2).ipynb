# Veri Yapıları Ödevi: Arama ve Sıralama Algoritmaları

```python
# --- ARAMA ALGORİTMALARI ---

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr, target):
    # Not: Binary Search için liste sıralı olmalıdır.
    low = 0
    high = len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

# --- SIRALAMA ALGORİTMALARI ---

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# --- TEST VE ÖRNEK KULLANIM ---

liste = [64, 34, 25, 12, 22, 11, 90]
print(f"Orijinal Liste: {liste}")

# Sıralama testleri
print(f"Bubble Sort: {bubble_sort(liste.copy())}")
print(f"Selection Sort: {selection_sort(liste.copy())}")
print(f"Merge Sort: {merge_sort(liste.copy())}")
print(f"Quick Sort: {quick_sort(liste.copy())}")

# Arama testleri
sirali_liste = sorted(liste)
print(f"\nSıralı Liste (Arama için): {sirali_liste}")
target = 22
print(f"Linear Search (Hedef {target}): Index {linear_search(sirali_liste, target)}")
print(f"Binary Search (Hedef {target}): Index {binary_search(sirali_liste, target)}")
```
