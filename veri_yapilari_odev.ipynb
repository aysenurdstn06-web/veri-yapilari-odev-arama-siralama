# Veri Yapıları Ödevi: Arama ve Sıralama Algoritmaları

```python
# ==========================================
# 1. ARAMA ALGORİTMALARI
# ==========================================

def linear_search(liste, hedef):
    """Her elemanı tek tek kontrol eden temel arama algoritması."""
    for i in range(len(liste)):
        if liste[i] == hedef:
            return i
    return -1

def binary_search(liste, hedef):
    """Sıralı dizide her adımda arama alanını yarıya indiren verimli algoritma."""
    dusuk = 0
    yuksek = len(liste) - 1
    while dusuk <= yuksek:
        orta = (dusuk + yuksek) // 2
        if liste[orta] == hedef:
            return orta
        elif liste[orta] < hedef:
            dusuk = orta + 1
        else:
            yuksek = orta - 1
    return -1

# ==========================================
# 2. SIRALAMA ALGORİTMALARI
# ==========================================

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        anahtar = arr[i]
        j = i - 1
        while j >= 0 and anahtar < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = anahtar
    return arr

def merge_sort(arr):
    if len(arr) > 1:
        orta = len(arr) // 2
        sol = arr[:orta]
        sag = arr[orta:]
        merge_sort(sol)
        merge_sort(sag)
        i = j = k = 0
        while i < len(sol) and j < len(sag):
            if sol[i] < sag[j]:
                arr[k] = sol[i]; i += 1
            else:
                arr[k] = sag[j]; j += 1
            k += 1
        while i < len(sol):
            arr[k] = sol[i]; i += 1; k += 1
        while j < len(sag):
            arr[k] = sag[j]; j += 1; k += 1
    return arr

# ==========================================
# 3. TEST VE ÇIKTILAR
# ==========================================

test_listesi = [64, 34, 25, 12, 22, 11, 90]
print(f"Orijinal Liste: {test_listesi}")
print(f"Sıralanmış Liste (Merge Sort): {merge_sort(test_listesi.copy())}")

# Arama Testi (Binary Search için sıralı liste gerekir)
sirali = sorted(test_listesi)
hedef = 22
print(f"Binary Search: {hedef} elemanı {binary_search(sirali, hedef)}. indekste.")
```
